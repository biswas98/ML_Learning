```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#giving name to each attribute
cols = ['fLength','fWidth','fSize','fConc','fConc1','fAsym','fM3Long','fM3Trans','fAlpha','fDist','class']

df = pd.read_csv('Magic Data/magic04.data' , names=cols)
```

```python
df
```

```python
df.shape # 'shape' function display the no. of rows (no. of data) & no. of columns (no. of attributes) = ( rows , cols )
```

```python
# converting the 'class' attribute into binary format i.e. 0 & 1 (g=1 , h=0)
df['class'] = (df['class'] == 'g').astype(int)
```

```python
df
```

```python
# plotting and comparing gamma & hadron visually
for label in cols[:-1]:
    plt.hist(df[df['class']==1][label], color='green', label='Gamma', alpha=0.7, density=True)
    plt.hist(df[df['class']==0][label], color='red', label='Hadron', alpha=0.7, density=True)
    plt.title(label)
    plt.ylabel('Probability')
    plt.xlabel(label)
    plt.legend()
    plt.show()
```
